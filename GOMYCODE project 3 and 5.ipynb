```python
import seaborn as sns
def plot_correlation_map(TT):

    corr = TT.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

### LOGISTIC REGRESSION

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.metrics import confusion_matrix
import matplotlib.mlab as mlab
%matplotlib inline
```

```python
cat_features = ["Sex"]

for cat_feature in cat_features:
  TT[f"{cat_feature}_cat"] = TT[cat_feature].astype('category')
  TT[f"{cat_feature}_cat"] = TT[f"{cat_feature}_cat"].cat.codes


TT.head()
```

```python
TT.drop(["Sex"], axis=1, inplace=True)
TT.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

#features extraction
x = TT[['Age', 'Sex_cat','Pclass']]
y = TT['Survived']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)  #splitting data with test size of 25%

logreg = LogisticRegression()   #build our logistic model
logreg.fit(x_train, y_train)  #fitting training data
y_pred  = logreg.predict(x_test)    #testing models performance
print("Accuracy={:.2f}".format(logreg.score(x_test, y_test)))
```
